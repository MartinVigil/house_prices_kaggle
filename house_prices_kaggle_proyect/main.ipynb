```python
#PREPROCESAMIENTO DE LOS DATOS
```

```python
train_df_c = train_df_c.drop(columns=[a for a in train_df_c.columns if (train_df_c[a].isna().sum()/train_df_c.shape[0]) > 0.20],axis=0)
train_df_c = train_df_c.drop(columns='Id')
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder

train_cat_cols = train_df_c.select_dtypes(include='O')
num_cols = train_df_c.select_dtypes(exclude='O')

prep_train = train_df_c.copy()

imputer_categorical = SimpleImputer(strategy='most_frequent')
prep_train[train_cat_cols.columns] = pd.DataFrame(imputer_categorical.fit_transform(prep_train[train_cat_cols.columns]))

prep_train.columns = train_df_c.columns

imputer_num = SimpleImputer()
prep_train[num_cols.columns] = pd.DataFrame(imputer_num.fit_transform(prep_train[num_cols.columns]))

prep_train.columns = train_df_c.columns

encoder = OrdinalEncoder()
prep_train[train_cat_cols.columns] = pd.DataFrame(encoder.fit_transform(prep_train[train_cat_cols.columns]))

prep_train.columns = train_df_c.columns

prep_train
```

```python
#normalizacion de las variables sesgadas
```

```python
from scipy import stats

norm = [x for x in prep_train[num_cols.columns] if abs(prep_train[x].skew()) > 0.5]

for x in prep_train[norm]:
    prep_train[x] = np.log1p(prep_train[x])    
```

```python
prep_train.hist(figsize=(20,20))
```

```python
#escalar
```

```python
from mlxtend.preprocessing import minmax_scaling

for x in prep_train.columns:
    prep_train[x] = minmax_scaling(prep_train,columns=[x])   
```

```python
from sklearn.model_selection import cross_val_score
from xgboost import XGBRegressor

y = prep_train['SalePrice']
features = [a for a in prep_train.columns if a != 'SalePrice']
x = prep_train[features]

model = XGBRegressor()
model.fit(x,y)

print(f'Precision: {cross_val_score(model,x,y,cv=5).mean()}')
```
